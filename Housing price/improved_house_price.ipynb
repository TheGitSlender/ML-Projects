```python
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
df.head()
```

```python
nan_col = [i for i in df.columns if df[i].isnull().any()]
for i in nan_col:
    print(i, df[i].isnull().sum())
```

```python
droppped_cols = ['Alley','MasVnrType','FireplaceQu','PoolQC','Fence','MiscFeature']
df = df.drop(droppped_cols,axis=1)
test = test.drop(droppped_cols,axis=1)
```

```python
df = df.apply(lambda x : x.fillna(x.value_counts().index[0]))
test = test.apply(lambda x : x.fillna(x.value_counts().index[0]))
```

```python
df.head()
```

```python
test = pd.get_dummies(test).astype(int)
df = pd.get_dummies(df).astype(int)
test.head()
```

```python
df.head()
```

```python
train_cols = df.columns
test_cols = test.columns

common_cols = train_cols.intersection(test_cols)
train_not_test = train_cols.difference(test_cols)
print(train_not_test)
```

```python
for i in train_not_test:
    test.insert(251,i,0)
test = test.drop(columns='SalePrice',axis=1)
```

```python
test.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=500)
```

```python
y = df['SalePrice']
X = df.drop('SalePrice',axis=1)
X = X.reindex(sorted(X.columns),axis=1)
```

```python
test = test.reindex(sorted(test.columns),axis=1)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,y_pred))
```

```python
model.fit(X,y)
```

```python
results = model.predict(test)
test['SalePrice'] = results
submission = test[['Id','SalePrice']]
submission.to_csv('submission.csv',index=False)
```
