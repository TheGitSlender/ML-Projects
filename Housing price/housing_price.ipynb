```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
df.head()
```

#### Null values

```python
nan_cols = [i for i in df.columns if df[i].isnull().any()]
test_nan_cols = [i for i in test.columns if df[i].isnull().any()]
```

```python
for i in nan_cols:
    print(i,df[i].isna().sum())
```

```python
dropped_features = ['Alley','PoolQC','Fence','MiscFeature','MasVnrType','FireplaceQu']
df = df.drop(dropped_features,axis=1)
test = test.drop(dropped_features,axis=1)
```

```python
#filling null values with most common instance
df = df.apply(lambda x: x.fillna(x.value_counts().index[0]))
test = test.apply(lambda x: x.fillna(x.value_counts().index[0]))
df
```

#### Analyze data

```python
df['SalePrice'].describe()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
```

```python
label_df = df.loc[:,(df=='Po').any()]
test_label_df = test.loc[:,(test=='Po').any()]
label_cols = label_df.columns
test_label_cols = test_label_df.columns
for i in label_cols:
    df[i] = lbl_encoder.fit_transform(df[i])
for i in test_label_cols:
    test[i] = lbl_encoder.fit_transform(test[i])
```

```python
df.describe()
```

```python
df = df.select_dtypes(include=['int64','float64'])
test = test.select_dtypes(include=['int64','float64'])
```

#### Find correlation between SalePrice and other features

```python
correlations = df.corr()
correlations = correlations["SalePrice"].sort_values(ascending=False) 
correlations
```

```python
y = df['SalePrice']
X = df.drop('SalePrice',axis=1)
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_squared_error
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100)
```

```python
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print(mean_squared_error(y_test,y_pred,squared=False))
```

```python
results = model.predict(test)
test['SalePrice'] = results

test
```

```python
submission = test[['Id','SalePrice']]
submission.to_csv('submission.csv')
```
