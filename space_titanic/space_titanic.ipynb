```python
#Imports that we need
import pandas as pd
```

```python
#We read the data and analyze it
df = pd.read_csv("train.csv")
test = pd.read_csv('test.csv')
print(df.shape)
df.head()
```

Preprocessing the data

```python
#Check for null values
df.isnull().sum()
```

```python
#Replace values by most common
df = df.apply(lambda x : x.fillna(x.value_counts().index[0]))
test = test.apply(lambda x : x.fillna(x.value_counts().index[0]))
```

```python
#Dropping features that do not affect the result
dropped_features = ['Name']
df = df.drop(dropped_features,axis=1)
test = test.drop(dropped_features,axis=1)
```

```python
#Preprocess categorical data
df['Transported'] = (df['Transported']==True).astype(int)
df['CryoSleep'] = (df['CryoSleep']==True).astype(int)
df['VIP'] = (df['VIP']==True).astype(int)


test['CryoSleep'] = (test['CryoSleep']==True).astype(int)
test['VIP'] = (test['VIP']==True).astype(int)
```

```python
#Separating the group data from the id of the person in 'PassengerId' column

df['col'] = df['PassengerId'].astype(str)
df['Group'] = df['col'].str[0:4]
df['Id'] = df['col'].str[5:7]

df.drop(['col','PassengerId'], axis=1, inplace=True)
#All steps repeated for the resulting dataset

test['col'] = test['PassengerId'].astype(str)
test['Group'] = test['col'].str[0:4]
test['Id'] = test['col'].str[5:7]

test.drop('col', axis=1, inplace=True)
pass_id = test.pop('PassengerId')
```

```python
df.head()
```

```python
df[['Deck','Num','Side']] = df['Cabin'].str.split('/', expand=True)
df.drop('Cabin',axis=1,inplace=True)
test[['Deck','Num','Side']] = test['Cabin'].str.split('/', expand=True)
test.drop('Cabin',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Group'] = df['Group'].astype(int)
df['Id'] = df['Id'].astype(int)
df['Num'] = df['Num'].astype(int)

test['Group'] = test['Group'].astype(int)
test['Id'] = test['Id'].astype(int)
test['Num'] = test['Num'].astype(int)
```

```python
df = pd.get_dummies(df).astype(int)
test = pd.get_dummies(test).astype(int)
df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
model = RandomForestClassifier(n_estimators=500)
```

```python
Y = df['Transported']
features = df[df.columns.drop('Transported')]
features.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(features,Y,test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
